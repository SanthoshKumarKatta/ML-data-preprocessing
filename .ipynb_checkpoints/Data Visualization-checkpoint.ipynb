```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_table('C:\Santhosh\AIML\Github\ML-data-preprocessing\data\Test_u94Q5KV.txt', delimiter=",", names=('Item_Identifier', 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility', 'Item_Type', 'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type'))
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def barGraph(feature):
    fig, ax = plt.subplots() 
    print(feature)
    data = df[feature].value_counts()
    x = data.index
    y= data.values
    ax.bar(x,y)
    ax.set_title('Category')
    ax.set_xlabel('Category/Group')
    ax.set_ylabel('Frequency')    
```

```python
def pieGraph(feature):
    # Data to plot
    labels = df[feature].value_counts().index
    sizes = df[feature].value_counts().values
    plt.pie(sizes, labels=labels)
    plt.axis('equal')
    plt.show()
```

```python
pieGraph('Item_Fat_Content')
pieGraph('Outlet_Size')
pieGraph('Item_Type')
pieGraph('Outlet_Type')
pieGraph('Outlet_Location_Type')
```

```python
barGraph('Item_Fat_Content')
barGraph('Outlet_Size')
barGraph('Item_Type')
barGraph('Outlet_Type')
barGraph('Outlet_Location_Type')
```
